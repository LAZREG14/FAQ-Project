```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from colorama import Fore, Style
import glob
import re
import pandas as pd
import numpy as np
import json
import time

def ouvrir_json(chemin):
    f = open(chemin, encoding="utf-8")
    toto = json.load(f)
    f.close()
    return toto
```

```python
start = time.time()

fichier = glob.glob("appariements_faq/*.json") #parcourir chaque fichier json
        #print(fichier)
for j in fichier:
    f=re.findall(r'\b\d[a-zA-Z_0-9]*', j) #récupérer le nom du fichier json
    print("-"*60)
    print("| fichier : {}|".format(f))
    print("-"*60)
    data= ouvrir_json(j)
    questions = []
    reponses = []
    for cle, valeur in data.items():
        questions.append(valeur[0]) #récupérer les questions et les stocker dans une liste
        reponses.append(valeur[1]) #récupérer les réponses et les stocker dans une autre liste
    index = 0 #création de l'index qui remplacera la clé de chaque question
    for question in questions:
        index += 1
        V = CountVectorizer() # préparer le vectoriseur
        qr_vecs = V.fit_transform([question]+reponses).toarray() # vectorisation des mots des deux listes ( questions & reponses)
        cos_sim = cosine_similarity([qr_vecs[0]],qr_vecs[1:]) #comparer la similarité cosinus entre une question et chaque réponse
        n=0
        while n <len(max(cos_sim)):
            if max(cos_sim)[n]==np.max(cos_sim):
                position=n+1
                if index == position:
                    b=(f'{Fore.GREEN}bonne{Style.RESET_ALL}')
                    print('La réponse {} est la plus proche de la question {}, avec un score de {} : la réponse est {}'.format(position,index,round(np.max(cos_sim),8),b))
                else:
                    m=(f'{Fore.RED}mauvaise{Style.RESET_ALL}')
                    print('La réponse {} est la plus proche de la question {}, avec un score de {} : la réponse est {}'.format(position,index,round(np.max(cos_sim),8),m))
                break
            else:
                 n=n+1                
end = time.time()
print("-"*60)
print(f"le temps d'exécution: {end - start} secondes")
```
