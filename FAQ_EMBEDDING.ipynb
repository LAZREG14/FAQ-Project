```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from colorama import Fore, Style
import json
import glob
import re
import numpy as np
import pandas as pd
import time


def ouvrir_json(chemin):
    f = open(chemin, encoding="utf-8")
    toto = json.load(f)
    f.close()
    return toto
```

```python
def similarité(model):
    fichier = glob.glob("appariements_faq/*.json")
        #print(fichier)
    for j in fichier:
        f=re.findall(r'\b\d[a-zA-Z_0-9]*', j)
        print("-"*60)
        print("| fichier : {}, model utilisé : {}|".format(f,model))
        print("-"*60)
        data= ouvrir_json(j)
        questions = []
        reponses = []
        for cle, valeur in data.items():
            questions.append(valeur[0])
            reponses.append(valeur[1])
        index = 0
        for question in questions:
            index += 1
            qr_vecs = SentenceTransformer(model).encode([question]+reponses)
            cos_sim = cosine_similarity ([qr_vecs[0]],qr_vecs[1:])
            n=0
            while n <len(max(cos_sim)):
                if max(cos_sim)[n]==np.max(cos_sim):
                    position=n+1

                    if index == position:
                        b=(f'{Fore.GREEN}bonne{Style.RESET_ALL}')
                        print('La réponse {} est la plus proche de la question {}, avec un score de {} : la réponse est {}'.format(position,index,round(np.max(cos_sim),8),b))
                    else:
                        m=(f'{Fore.RED}mauvaise{Style.RESET_ALL}')
                        print('La réponse {} est la plus proche de la question {}, avec un score de {} : la réponse est {}'.format(position,index,round(np.max(cos_sim),8),m))
                    break
                else:
                     n=n+1
```

```python
start = time.time()
similarité('all-mpnet-base-v2')
end = time.time()
print("-"*60)
print(f"le temps d'exécution: {end - start} secondes")
```
